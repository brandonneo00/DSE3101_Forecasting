```python
# Path to the output text file
output_file = "impt.txt"

def convert_to_str(value):
    if isinstance(value, np.float64):
        return str(value)
    return value

# Write the data to a text file
with open(output_file, "w") as file:
    for sublist in impt:
        print(len(sublist))
        sublist_str = [convert_to_str(item) for item in sublist]
        file.write(", ".join(sublist_str) + "\n")
```
